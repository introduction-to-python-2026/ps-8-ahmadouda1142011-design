```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv"
df = pd.read_csv(url)

# Save CSV
df.to_csv("usa_states_2014.csv", index=False)

# Select columns
states_df = df[["State", "Population", "Rank"]]

# Visualization
top10 = states_df.sort_values("Population", ascending=False).head(10)

plt.figure()
plt.bar(top10["State"], top10["Population"])
plt.xlabel("State")
plt.ylabel("Population")
plt.title("Top 10 Most Populous US States (2014)")
plt.xticks(rotation=45)
plt.show()
```
